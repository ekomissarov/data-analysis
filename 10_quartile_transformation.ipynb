```python
import numpy as np
import scipy.stats as sts
import seaborn as sns

import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
uniform_rv = sts.uniform(loc=0, scale=1)
y = uniform_rv.rvs(10**6)
sns.histplot(y)
```

```python
# cumulative distribution function of an exponential random variable: 1-exp(alpha*x)
# cdf^-1: -(1/alpla) * ln(1-y)

x = -0.5*np.log(1 - y)
sns.histplot(x, stat='density');
```

```python
#now let's try in the opposite direction
exponential_rv = sts.expon(scale=0.5)
x = exponential_rv.rvs(10**6)
sns.histplot(x)
```

```python
y = 1 - np.exp(-2*x)
sns.histplot(y, stat='density');
```
